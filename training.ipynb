````python
df_model = df[
    ['Płeć', 'Rocznik', '5 km Czas', 'Czas']
].copy()

df_model = df_model.dropna()

print(df_model.shape)
````

````python
df_model.isna().sum()
````

````python
df_model[['5 km Czas', 'Czas']].head(10)
````

````python
df_model["5km_seconds"] = (
    pd.to_timedelta(df_model["5 km Czas"])
    .dt.total_seconds()
)

df_model["halfmarathon_seconds"] = (
    pd.to_timedelta(df_model["Czas"])
    .dt.total_seconds()
)

df_model[
    ["5 km Czas", "5km_seconds",
     "Czas", "halfmarathon_seconds"]
].head()
````

````python
CURRENT_YEAR = 2024

df_model["wiek"] = CURRENT_YEAR - df_model["Rocznik"]
````

````python
df_model[["Rocznik", "wiek"]].head()
````

````python
X = df_model[
    ["Płeć", "wiek", "5km_seconds"]
]

y = df_model["halfmarathon_seconds"]
````

````python
df_model[
    ["wiek", "5km_seconds", "halfmarathon_seconds"]
].describe()
````

````python
df_model = df_model[
    (df_model["wiek"] >= 18) &
    (df_model["wiek"] <= 90) &
    (df_model["5km_seconds"] > 0)
].copy()

print(df_model.shape)
````

````python
df_model[
    ["wiek", "5km_seconds", "halfmarathon_seconds"]
].describe()
````

````python
df_model["gender"] = df_model["Płeć"].map({
    "M": 1,
    "K": 0
})
````

````python
df_model[["Płeć", "gender"]].head()
````

````python
model_data = df_model[
    [
        "gender",
        "wiek",
        "5km_seconds",
        "halfmarathon_seconds"
    ]
].copy()
````

````python
model_data.head()
````

````python
from pycaret.regression import *
````

````python
reg_setup = setup(
    data=model_data,
    target="halfmarathon_seconds",
    session_id=42,
    verbose=False
)
````

````python
best_model = compare_models()
````

````python
final_model = finalize_model(best_model)
````

````python
save_model(
    final_model,
    "models/halfmarathon_model"
)
````

````python
from openai import OpenAI
from dotenv import load_dotenv
import os
import json

load_dotenv(override=True)

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

user_text = """
Jestem kobietą, mam 38 lat,
5 km biegam w 18 minut
"""

response = client.responses.create(
    model="gpt-4.1-mini",
    input=f"""
Wyciągnij dane z tekstu.

Zwróć WYŁĄCZNIE JSON.

Format:

{{
  "gender": 0,
  "wiek": 0,
  "5km_seconds": 0
}}

Zasady:
- kobieta = 0
- mężczyzna = 1
- czas podaj w sekundach

Tekst:
{user_text}
"""
)

print(response.output_text)
````

````python
import json
import pandas as pd
from pycaret.regression import predict_model

clean_json = (
    response.output_text
    .replace("```json", "")
    .replace("```", "")
    .strip()
)

data = json.loads(clean_json)

test_runner = pd.DataFrame({
    "gender": [data["gender"]],
    "wiek": [data["wiek"]],
    "5km_seconds": [data["5km_seconds"]]
})

prediction = predict_model(
    final_model,
    data=test_runner
)

prediction[["prediction_label"]]
````
